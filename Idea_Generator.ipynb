```
import pickle
import pandas as pd
import numpy as np
import keras
from sklearn.model_selection import train_test_split
from keras.models import Sequential
from keras.layers import Dense, Dropout, LSTM, Embedding
from keras.optimizers import Adam
from keras.preprocessing.text import Tokenizer
from keras.utils.np_utils import to_categorical  
```

**For Sentences**

```
with open('/content/gdrive/My Drive/iemocap/train/sentences.pkl', 'rb') as f:
    data = pickle.load(f)
print(data)
```

**For Labels**

```
with open('/content/gdrive/My Drive/iemocap/train/labels.pkl', 'rb') as f:
    labels = pickle.load(f)

print(labels)
```

```
processed_data=[]
processed_label=[]
for i in range(0,len(data)):
  for j in range(0,len(data[i])):
    text=data[i][j]
    label=labels[i][j]
    processed_data.append(text)
    processed_label.append(label)
```

```
t = Tokenizer()
t.fit_on_texts(processed_data)
X = t.texts_to_matrix(processed_data, mode='tfidf')
Y = to_categorical(processed_label, num_classes=6)
```

```
lstm_out = 200
embed_dim=128
model = Sequential()
model.add(Embedding(2500, embed_dim,input_length = X.shape[1]))
model.add(LSTM(lstm_out))
model.add(Dense(2,activation='softmax'))
model.compile(loss = 'categorical_crossentropy', optimizer='adam',metrics = ['accuracy'])
print(model.summary())
```

```
X_train, X_valid, Y_train, Y_valid = train_test_split(X,Y, test_size = 0.20, random_state = 36)
```

```
model.fit(X_train, Y_train, epochs = 10,  verbose = 5)
```
